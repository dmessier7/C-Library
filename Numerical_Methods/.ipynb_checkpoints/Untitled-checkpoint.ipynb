```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = [[-3.14159265358979, 0.00000000000000],
    [-2.81089869005402, -0.43702230525987],
    [-2.48020472651826, -0.78170987707147],
    [-2.14951076298249, -0.85207605602719],
    [-1.81881679944672, -0.52335935447908],
    [-1.48812283591095, 0.22121102681686],
    [-1.15742887237519, 1.27130734031775],
    [-0.82673490883942, 2.41089249371528],
    [-0.49604094530365, 3.37539264122665],
    [-0.16534698176788, 3.92749194128624],
    [0.16534698176788, 3.92749194128624],
    [0.49604094530365, 3.37539264122665],
    [0.82673490883942, 2.41089249371528],
    [1.15742887237519,1.27130734031775],
    [1.48812283591095, 0.22121102681686],
    [1.81881679944672, -0.52335935447908],
    [2.14951076298249, -0.85207605602719],
    [2.48020472651826, -0.78170987707147],
    [2.81089869005402, -0.43702230525987],
    [3.14159265358979, 0.00000000000000]]

def extract(lst, num):
    n = []
    for i in lst:
        #print(i[0])
       n.append(i[num])
    return n

X = extract(T, 0)
Y = extract(T, 1)

def poly(x):
    y = 3.4950 + (-1.5285*(x**2)) +(0.1233*(x**4))
    return y

def trig(x):
    y = (0.8612) + (2.1258*np.cos(x))+ (0.9097*np.cos(2*x))+(-0.1393*np.cos(3*x))
    return y

Xfunc = numpy.arange(-3.5,3.5,0.0001)
Ypoly = []
for i in Xfunc:
    Ypoly.append(poly(i))
Ytrig = []
for i in Xfunc:
    Ytrig.append(trig(i))
```

```python
plt.scatter(X,Y)
plt.plot(Xfunc,Ypoly)
plt.suptitle("4th Degree Polynomial Approximation")
plt.title("P4(x) = 3.4950 + 0.0000x -1.5285 x^2 + 0.0000x^3 + 0.1233x^4")
plt.ylabel('y')
plt.xlabel('x')
plt.savefig('Poly.pdf', bbox_inches='tight',pad_inches=1.0)
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(Xfunc,Ytrig)
plt.title("S3(x)=0.8612-0.1393Cos3x+2.1258Cosx+0.9097Cos2x+ 0.0000Sinx+ 0.0000Sin2x")
plt.suptitle("Trig Polynomial Approximation")
plt.ylabel('y')
plt.xlabel('x')
plt.savefig('Trig.pdf', bbox_inches='tight',pad_inches=1.0)
plt.show()
```
